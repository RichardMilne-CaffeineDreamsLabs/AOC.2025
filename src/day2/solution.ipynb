```python
input = open("input.txt")
prids = input.read().split(",")
prids_expanded = [list(range(int(s),int(e)+1)) for s, e in (prid.split("-") for prid in prids)]

#print(prids_expanded)
input.close()
```

```python
import re

naughty_list = list[int]()
pattern = r"^(?:(\d+?)\1{1,})$"

for prid_range in prids_expanded:
    #naughty_prids = [p for p in prid_range if re.search(str(p), pattern) is not None]
    for p in prid_range:
        match = re.search(pattern, str(p))
        if match is not None:
            # is_naughty = len(match.group(0)) % 2 == 0
            naughty_list.append(p)

naughty_sum = sum(naughty_list)
print(naughty_sum)
```
