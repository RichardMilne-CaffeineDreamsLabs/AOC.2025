```julia
file = open("input.txt", "r");
lines = readlines(file);
min = 0
max = 99

r = min:max

range = collect(r)
n = max - min + 1  
start = 50
count = 0
wrapsTotal = 0
for ln in lines
    ln = strip(ln)
    isempty(ln) && continue

    previous_start = start

    op = ln[1]
    val = parse(Int, ln[2:end])

    wraps = 0
    if op == 'R'
        new_position = previous_start + val
        wraps = div(new_position, n)
        start = mod(new_position, n)
    elseif op == 'L'
        new_position = previous_start - val
        s = previous_start == 0 ? n : previous_start
        wraps = val < s ? 0 : div(val - s, n) + 1
        start = mod(new_position, n)
    end

    wrapsTotal += wraps

    if start == 0
        count += 1
    end
end
println("Count of times at position 0:\t", count)
println("Total passing or at position 0:\t", wrapsTotal)

close(file);
```
