```julia
file = open("input.txt", "r");
lines = readlines(file);
min = 0
max = 99

r = min:max
```

```julia
range = collect(r)
n = max - min + 1  

start = 50

count = 0
wrapsTotal = 0
for ln in lines
    ln = strip(ln)
    isempty(ln) && continue

    previous_start = start

    op = ln[1]
    val = parse(Int, ln[2:end])

    wraps = 0
    if op == 'R'
        # full forward wraps = floor((previous + val) / n)
        wraps = div(previous_start + val, n)
        start = mod(previous_start + val, n)
    elseif op == 'L'
        # full backward wraps = ceil((val - previous) / n) when val > previous
        wraps = val <= previous_start ? 0 : div(val - previous_start + n - 1, n)
        start = mod(previous_start - val, n)

    end

    if start == 0
        count += 1
    else
        wrapsTotal += wraps
    end
end
println("Count of times at position 0: ", count)
println("Total wraps around the range: ", wrapsTotal)

close(file);
```
