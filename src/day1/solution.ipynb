```julia
file = open("input.txt", "r");
lines = readlines(file);
min = 0
max = 99

r = min:max

range = collect(r)
n = max - min + 1  
start = 50
count = 0
wrapsTotal = 0
for ln in lines
    ln = strip(ln)
    isempty(ln) && continue

    previous_start = start

    op = ln[1]
    val = parse(Int, ln[2:end])

    wraps = 0
    fwd_wraps = 0
    bwd_wraps = 0
    if op == 'R'
        # full forward wraps = floor((previous + val) / n)
        new_position = previous_start + val
        
        start = mod(new_position, n)

        # fwd_wraps = start == 0 ? 0 : floor(new_position / n)
        fwd_wraps = start == 0 ? 0 : div(new_position, n)
        wraps += fwd_wraps
        #wraps = div(previous_start + val, n)
    elseif op == 'L'
        # full backward wraps = ceil((val - previous) / n) when val > previous
        
        new_position = previous_start - val
        #bwd_wraps = abs(floor(new_position / n))
        start = mod(new_position, n)
        # bwd_wraps = start == 0 ? 0 : val < previous_start ? 1 : div(val - previous_start + n - 1, n)
        bwd_wraps = start == 0 ? 0 : val > previous_start ? ceil((val - previous_start) / n) : 0
        wraps += bwd_wraps
        

    end

    if start == 0
        count += 1
    else
        wrapsTotal += wraps
    end
end
println("Count of times at position 0:\t", count)
println("Total wraps around the range:\t", wrapsTotal)
println("Total times at or passing 0:\t", count + wrapsTotal)

@show fwd_wraps
@show bwd_wraps

close(file);
```
